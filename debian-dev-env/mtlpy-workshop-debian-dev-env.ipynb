## Debian : Installing a Debian Package

How to find a package

* Web : all packages : [http://packages.debian.org/stable/allpackages](http://packages.debian.org/en/stable/allpackages)
* Web : all packages by category : [http://packages.debian.org/stable/](http://packages.debian.org/en/stable/)
* Web : Python packages : [http://packages.debian.org/stable/python/](http://packages.debian.org/en/stable/python/)

How to install (and remove) a package

* Graphic mode : where to go<pre>
Applications > System Tools > Add/Remove Softwares
</pre>
* Command line : here's the command<br />
(replace "package-name" with the actual package name you want to install/remove)<pre>
$ sudo apt-get install package-name
$ sudo apt-get remove package-name
</pre>

Packages installed system-wide vs installed project-wide

* System-wide : for all projects, might need "sudo"
* Project-wide : project specific, permits specific versions

* Why control versions of source code?

## Python Extensions

* Python extensions written in other languages might be needed, so we need tools to compile them<br />
ex.: gcc, pil...
* build-essentials, python-dev (Python sources)
* Debian packages = 
<pre>
$ sudo apt-get install build-essential
$ sudo apt-get install python-dev
</pre>
